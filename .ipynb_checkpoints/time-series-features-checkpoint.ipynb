```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'/Users/zoeyneo/code/zoeyneo/investoinsight/first_75_tickers.csv')
```

```python
df.head()
```

```python
df['fiscalDateEnding'] = pd.to_datetime(df['fiscalDateEnding'])
df['fiscalDateEnding'].info()
```

```python
#duplicate values for every month
df = df.set_index(df['fiscalDateEnding'])

def reindex_by_date(df):
    dates = pd.date_range(df.index.min(), df.index.max(),freq="MS")
    return df.reindex(dates).ffill()

df = df.groupby(df['Symbol']).apply(reindex_by_date).reset_index(0, drop=True)
```

```python
df[df['Symbol']=="MMM"].head(20)
```

```python
df['shortTermDebt'] = pd.to_numeric(df['shortTermDebt'], errors='coerce').convert_dtypes()
df['longTermDebt'] = pd.to_numeric(df['longTermDebt'], errors='coerce').convert_dtypes()
# print(df['shortTermDebt'].dtype)
# print(df['longTermDebt'].dtype)

#solvency ratio
df['debt_to_equity'] = (df['shortTermDebt'] + df['longTermDebt']) / df['totalShareholderEquity']
# print(df['debt_to_equity'].dtype)

#profitability
# df['pe_ratio'] = 
```
