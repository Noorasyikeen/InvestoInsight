```python
#duplicate values for every month


def reindex_by_date(df):
    df.drop_duplicates(subset='fiscalDateEnding', keep='first', inplace=True)
    
    dates = pd.date_range('2018-11-01', '2022-01-01', freq="MS")
    columns_to_fill = df.columns.difference(['Date'])
    df = df.set_index('fiscalDateEnding').reindex(dates)

    # Forward-fill columns within each group
    df['totalRevenue'].ffill(inplace=True)
    df['Symbol'].ffill(inplace=True)
    df['shortTermDebt'].ffill(inplace=True)
    df['longTermDebt'].ffill(inplace=True)
    df['totalShareholderEquity'].ffill(inplace=True)
    df['commonStockSharesOutstanding'].ffill(inplace=True)
    df['netIncome'].ffill(inplace=True)
    df['ebitda'].ffill(inplace=True)
    df['reportedCurrency_y'].ffill(inplace=True)
    
    # Reindex the DataFrame with the specified date range
#     df = df.set_index('fiscalDateEnding').reindex(dates)
    
    # Backward-fill any remaining missing values
    df['totalRevenue'].bfill(inplace=True)
    df['Symbol'].bfill(inplace=True)
    df['shortTermDebt'].bfill(inplace=True)
    df['longTermDebt'].bfill(inplace=True)
    df['totalShareholderEquity'].bfill(inplace=True)
    df['commonStockSharesOutstanding'].bfill(inplace=True)
    df['netIncome'].bfill(inplace=True)
    df['ebitda'].bfill(inplace=True)
    df['reportedCurrency_y'].bfill(inplace=True)
    
    return df


# Group by 'Symbol' and apply the reindexing function to each group
monthly_df = df.groupby(df['Symbol']).apply(reindex_by_date)
# .reset_index(0, drop=True)

monthly_df.head(50)
```
